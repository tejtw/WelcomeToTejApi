```python
# 導入TEJ API的套件
import tejapi
import pandas as pd

# 設定全域變數-登入TEJ API
tejapi.ApiConfig.api_key = 'your_key'
# 決定時間欄位是否顯示時區
tejapi.ApiConfig.ignoretz = True
#False

# 使用資料庫讀取所有資料，要習慣加入'paginate=True'來讀取所有資料
data = tejapi.get('TRAIL/TAPRCD'
                 #,paginate=True
                 )
# 展示資料庫的資料
data.head(5)
```

## **1.抓取資料庫中所有資料**
[試用的股票資料庫](https://api.tej.com.tw/columns.html?idCode=TRAIL/TAPRCD)

```python
# 使用資料庫讀取所有資料，要習慣加入'paginate=True'
data = tejapi.get('TRAIL/TAPRCD'
                 #,paginate=True
                 )

# 展示資料庫的資料
print(data)
```

## **2.抓取資料庫中特定欄位資料**
+ 以股票欄位、時間欄位做說明

```python
# 使用資料庫讀特定的股票資料，我們可以指定要coid欄位為'股票代碼'
# 要選取多個資料可以使用list的方式選取多筆股票['代號','代號']

tax = ['2330','0050','00878']
data = tejapi.get('TRAIL/TAPRCD',
                  coid= tax,
                  mdate='2022-01-03',
                  paginate=True)

# 展示資料庫的資料
print(data)
```

## **3.使用opts來塞選出指定的資料欄位**

```python
# 使用可以加入opts的字典函數來塞選需要的資料欄位的
#這次選取的columns有包含股票代碼、日期、開盤價
data = tejapi.get('TRAIL/TAPRCD',
                  coid='0050',
                  mdate='2022-01-03',
                  opts={'columns':['coid','mdate','open_d']},
                  paginate=True)

# 展示資料庫的資料
print(data)
```

## **4.使用mdate中的gte、lte指定資料期間**

```python
# 使用mdate字典函數來選定指定日期
data = tejapi.get('TRAIL/TAPRCD',
                  coid='0050',

                  #起始日期 gte = greater than even
                  #結束日期 lte = less than even
                  mdate={'gte':'2022-01-03','lte':'2022-02-28'},
                  opts={'columns':['coid','mdate','open_d']},
                  paginate=True)

# 展示資料庫的資料
print(data)
```

## **5.資料欄位改成中文顯示**

```python
data = tejapi.get('TRAIL/TAPRCD',
                  coid='0050',
                  mdate={'gte':'2022-01-03','lte':'2022-02-30'},
                  opts={'columns':['coid','mdate','open_d']},
                  paginate=True,
                  # 這邊加入 chinese_column_name=True讓欄位名稱變成中文版本
                  chinese_column_name=True
                 )

# 展示資料庫的資料
print(data)
```

# **免費版與付費版的資料庫差異**

### **付費版本的資料庫**
   + 怎麼購買Tejapi的付費資料庫
   + [TEJAPI Eshop資料庫](https://eshop.tej.com.tw/E-Shop/Edata_intro)
   + 付費版本的資料庫特點:
    1. 可以依據自己的需求去挑選需要的資料範圍
    2. 每日的資料呼叫次數最多呼叫1000次
    3. 最高每天可以呼叫1,000,000筆資料
    4. 總資料筆數最高達到11,747,555筆
    5. 資料年期為本日日期回推5年前的資料
    

### **免費版本的資料庫**
   + 試用版本的資料庫
   + [TEJAPI 試用版資料庫](https://api.tej.com.tw/datatables.html?db=TRAIL&t=%E8%A9%A6%E7%94%A8%E8%B3%87%E6%96%99%E5%BA%AB)
   + 免費版本的資料庫特點:
     1. 只能使用回推1年前的資料
     2. 每日的資料呼叫數量最多呼叫50000筆
     3. 每日的資料呼叫次數最多呼叫500次

## **3.導入套件**

```python
# 導入TEJ API的套件
import tejapi
# 設定全域變數-登入TEJ API
api_key = 'your_key'
tejapi.ApiConfig.api_key = api_key
# 決定時間欄位是否顯示時區
tejapi.ApiConfig.ignoretz = True
```

## **4. 從付費財務資料庫撈取資料**
### **4.1撈取台積電所有季度財務數據**

```python
# 因為是付費資料庫，固有所有資料，且資料不用另外調整
finance = tejapi.get('TWN/EWIFINQ',
                  coid='2330',
                  paginate=True,
                  chinese_column_name=True
                 )
finance
```

### **4.2 撈取多家公司所有季度、特定財務數據**

```python
# 挑選以下幾間公司
coid_list = ['2330','2311','2329','2331','2338']

coid_finance = tejapi.get('TWN/EWIFINQ',
                  coid= coid_list,
                  paginate=True,
                  chinese_column_name=True
                 )
coid_finance
```

```python
#查看資料欄位
coid_finance.columns
```

```python
#塞選出特定欄位
selected_finance = coid_finance[['證券碼','財務資料日','ROA(C) 稅前息前折舊前','營業毛利率','流動資產','現金流量比率']]
selected_finance
```

## **5.常用見的財務數據操作**

### **5.1使用sort_values()排序**

```python
#按照日期排序，使用sort_values()就可以讓資料由小到大排序
sorted_finance = selected_finance.sort_values(by = '財務資料日')
sorted_finance
```

## **5.2使用reset_index(drop=true)重置索引**

```python
#因為排序、塞選、刪除部分資料後索引值都會被打亂，所以要重置
final = sorted_finance.reset_index(drop = True)
final
```
